```python
from typing_extensions import final
import requests
from bs4 import BeautifulSoup
import csv
```

Ketiga, buat function untuk mengambil data dari websitenya

```python
def get_data (url):
  page = requests.get(url)
  if page.status_code == 200:
    page_html = page.content
    page_soup = BeautifulSoup(page_html, 'html.parser')
    data = []

    container_header = page_soup.find('div', {'style' : 'float:left; width:540px;'})
    judul = container_header.find('a', class_='title').text.strip().replace('\r\n', '')
    person = container_header.find_all('span')
    abstrak = page_soup.find('p', {'align' : 'justify'}).text.strip().replace('\r\n', '')
    data.append(judul)
    for i in person:
      split_text = i.text.strip().split(':')
      data.append(split_text[1])
    data.append(abstrak)
    return (data)
```

Fungsi `get_data` ini mengambil data dari sebuah halaman web dengan URL yang diberikan. Pertama, halaman web diunduh menggunakan `requests.get(url)`. Jika status kode respons adalah 200 (berarti halaman berhasil diunduh), maka kontennya diambil dan dianalisis menggunakan BeautifulSoup dengan parser HTML bawaan.

Data yang diambil termasuk judul, informasi mengenai orang (misalnya, penulis atau pembimbing), dan abstrak. Kontainer untuk judul dan informasi orang diidentifikasi dengan mencari elemen HTML yang sesuai, seperti `container_header` untuk judul. Setelah itu, data diekstrak dan dimasukkan ke dalam list `data`.

Pada bagian judul, elemen `a` dengan kelas `title` diambil, dan teksnya diambil, di-strip, dan beberapa karakter \r\n yang mungkin ada di dalamnya dihapus. Untuk informasi orang, elemen `span` diambil dan diiterasi. Setiap teks span di-split berdasarkan tanda titik dua (:) dan nilai kedua dari hasil split (indeks 1) ditambahkan ke dalam list data.

Abstrak diambil dengan mencari elemen `p` yang memiliki atribut `align` bernilai 'justify'. Kemudian teksnya diambil, di-strip, dan beberapa karakter \r\n dihapus.

Semua data yang telah diekstrak kemudian dimasukkan ke dalam list `data`, dan list tersebut dikembalikan sebagai output dari fungsi.

Keempat, buat function untuk mengambil link selengkapnya dan juga menyimpan data

```python
def crawl_pta_trunojoyo(url):
  response = requests.get(url)

  if response.status_code == 200:
    html = response.content
    soup = BeautifulSoup(html, 'html.parser')
    next_button = soup.find_all('a', class_='pag_button')
    for i in next_button:
      if i.text.strip() == '>' :
        next_page_link = i.get('href')
    data = []


    content_page = soup.find('ul', class_='items')
    link_content = content_page.find_all('li')
    for detail in link_content:
      link_view = detail.find('a', class_='gray button').get('href')
      data.append(get_data(link_view))
    return(data, next_page_link)
```

Fungsi `crawl_pta_trunojoyo` ini digunakan untuk melakukan crawling data dari halaman web dengan URL tertentu. Pertama, fungsi melakukan permintaan HTTP ke URL yang diberikan menggunakan `requests.get(url)`. Jika status kode respons adalah 200 (berhasil), maka kontennya diambil dan dianalisis menggunakan BeautifulSoup dengan parser HTML bawaan.

Selanjutnya, fungsi mencari semua elemen `a` dengan kelas `pag_button`, yang mungkin merupakan tombol navigasi ke halaman berikutnya. Iterasi dilakukan untuk mencari tombol navigasi yang memiliki teks '>' (kanan). Jika ditemukan, link untuk halaman berikutnya diambil dari atribut `href` dan disimpan dalam variabel `next_page_link`.

Selanjutnya, fungsi membuat list kosong bernama `data`. Data akan diisi dengan informasi yang diambil dari setiap elemen `li` yang berada di dalam elemen `ul` dengan kelas `items`. Setiap elemen `li` berisi link yang akan diambil menggunakan `get_data`, dan hasilnya dimasukkan ke dalam list `data`.

Terakhir, fungsi mengembalikan tuple yang berisi list `data` dan link untuk halaman berikutnya (`next_page_link`). Jika tidak ada link halaman berikutnya, `next_page_link` akan bernilai `None`.

Kelima, membuat sebuah fungsi main untuk memanggil berapa page untuk crawling data dan code untuk membuka file setelah berhasil di crawl

```python
def main(url, num_page):
  next_page_link = url
  final_crawl = []
  for i in range(num_page):
    hasil_crawl, next_page_link = crawl_pta_trunojoyo(next_page_link)
    final_crawl+=hasil_crawl

  # Membuka file CSV untuk penulisan
  with open('pta_trunojoyo_tugas.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['Judul', 'Penulis', 'Dosen Pembimbing I', 'Dosen Pembimbing II', 'Abstrak'])
    csvwriter.writerows(final_crawl)
  print(f'Data Berhasil Di Simpan Dengan Jumlah {len(final_crawl)} data')
```

Fungsi `main` ini bertujuan untuk melakukan crawling sejumlah halaman tertentu dari situs web dengan URL awal yang diberikan (`url`). Jumlah halaman yang akan dicrawl ditentukan oleh parameter `num_page`. Fungsi ini menggunakan fungsi `crawl_pta_trunojoyo` untuk mendapatkan data dari setiap halaman, dan hasilnya disimpan dalam list `final_crawl`.

Fungsi menggunakan loop `for` untuk iterasi sebanyak `num_page`. Pada setiap iterasi, fungsi `crawl_pta_trunojoyo` dipanggil dengan menggunakan link halaman berikutnya (`next_page_link`). Hasil crawling dan link untuk halaman berikutnya kemudian ditambahkan ke dalam list `final_crawl` menggunakan operator `+=`.

Setelah selesai iterasi, fungsi membuka file CSV (`pta_trunojoyo_tugas.csv`) untuk penulisan data. Header CSV ditulis terlebih dahulu menggunakan `csvwriter.writerow(['Judul', 'Penulis', 'Dosen Pembimbing I', 'Dosen Pembimbing II', 'Abstrak'])`. Kemudian, seluruh data yang telah dicrawl (`final_crawl`) ditulis ke dalam file CSV menggunakan `csvwriter.writerows(final_crawl)`.

Terakhir, fungsi mencetak pesan bahwa data telah berhasil disimpan bersama dengan jumlah data yang berhasil diambil (`len(final_crawl)`).

 1. **Menghapus tanda baca**

  Salah satu langkah umum dalam preprocessing teks adalah menghapus tanda baca. Langkah ini bertujuan untuk membersihkan teks dari elemen-elemen yang tidak relevan agar analisis atau pemodelan dapat dilakukan dengan lebih akurat.

```python
import re
def cleaning(data):
  data = re.sub(r'#[A-Za-z0-9]+',' ',data)
  data = re.sub(r"http\S+",' ',data)
  data = re.sub(r'[0-9]+',' ',data)
  data = re.sub(r"[-()\"#/@;:<>{}'+=~|.!?,_]", " ", data)
  data = data.strip(' ')
  data = data.strip("\n")
  return data

df["cleaning"]= df["Abstrak"].apply(cleaning)
df.head()
```

code diatas bertujuan untuk membersihkan teks dalam kolom "Abstrak" dari DataFrame df dari elemen-elemen seperti hashtag, URL, angka, karakter khusus, dan spasi ekstra. Hasil pembersihan disimpan dalam kolom baru bernama "cleaning".

# 2. Case Folding (Mengubah huruf kapital menjadi huruf kecil)

  Case Folding adalah proses mengubah semua huruf kapital menjadi huruf kecil dalam teks. Misalnya, jika kita memiliki kata "HELLO" setelah Case Folding, kata tersebut akan menjadi "hello". Tujuannya adalah untuk membuat teks lebih seragam sehingga komputer dapat lebih mudah memahami dan memproses informasi.

```python
df['casefolding'] = [entry.lower() for entry in df['cleaning']]

df.head(5)
```

# **Term Frequency**

Term Frequency (TF) atau Frekuensi Kata mengukur seberapa sering sebuah kata muncul dalam sebuah dokumen. Konsep ini adalah bagian penting dari analisis teks dan pemrosesan bahasa alami. Term Frequency dapat membantu dalam menentukan seberapa penting sebuah kata dalam suatu dokumen.

Rumus Term Frequency (TF) pada umumnya dinyatakan sebagai jumlah kemunculan kata (term) dalam suatu dokumen dibagi dengan total jumlah kata dalam dokumen tersebut. Terdapat beberapa variasi rumus TF, dan salah satu yang umum digunakan adalah sebagai berikut:

$\text{TF}(t, d) = \frac{\text{jumlah kemunculan term } t \text{ dalam dokumen } d}{\text{total jumlah term dalam dokumen } d}$








```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
df['Stopword Removal'] = df['Stopword Removal'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
vectorizer = CountVectorizer()
count_matrix = vectorizer.fit_transform(df['Stopword Removal'])
count_array = count_matrix.toarray()
corpus_vsm = pd.DataFrame(data=count_array,columns = vectorizer.vocabulary_.keys())
new_column = df['Stopword Removal']
corpus_vsm.insert(0, 'Teks', new_column)
corpus_vsm
```

code diatas digunakan dalam proses pengolahan teks menggunakan Python, terutama dengan menggunakan library Pandas dan Scikit-learn. Pada tahap awal, langkah-langkah dilakukan untuk menggabungkan kata-kata dalam daftar menjadi satu string dalam kolom 'Stopword Removal'. Selanjutnya, library CountVectorizer dari Scikit-learn digunakan untuk mengonversi teks ke dalam matriks frekuensi kata, di mana setiap baris mewakili dokumen dan setiap kolom mewakili kata unik. Hasilnya kemudian disimpan dalam bentuk DataFrame baru, corpus_vsm, yang mencakup matriks frekuensi kata dan kolom 'Teks' yang berisi teks asli sebelum penghapusan stopword.

  Logarithmic Frequency (Frekuensi Logaritmik) adalah sebuah metode normalisasi yang umumnya digunakan dalam pemodelan teks untuk mengatasi masalah ketidakseimbangan frekuensi kata. Metode ini bertujuan untuk memberikan bobot yang lebih seimbang pada kata-kata yang muncul dalam jumlah yang beragam di dalam dokumen. Dengan kata lain, kata-kata yang muncul sangat sering akan memiliki dampak yang lebih terukur.

  Proses Logarithmic Frequency melibatkan penggunaan fungsi logaritma natural terhadap jumlah kemunculan suatu term dalam dokumen. Rumusnya dapat dinyatakan sebagai berikut:

  $ \text{Logarithmic Frequency}(t, d) = 1 + \log(\text{jumlah kemunculan term } t \text{ dalam dokumen } d)$

  
  Penggunaan logaritma memberikan dampak yang lebih moderat terhadap kata-kata yang muncul lebih sering. Jumlah kemunculan yang tinggi akan memberikan nilai logaritma yang lebih besar, tetapi dampaknya tidak langsung proporsional dengan jumlah kemunculan. Ini membantu mengurangi efek skewness (ketidakseimbangan) dalam frekuensi kata-kata.

  Metode Logarithmic Frequency sering digunakan dalam skema pembobotan seperti TF-IDF, di mana representasi vektor dokumen diperoleh dengan memperhitungkan frekuensi kata dan bobotnya.


```python
vectorizer = CountVectorizer()
count_matrix = vectorizer.fit_transform(df['Stopword Removal'])
log_count_matrix = np.log1p(count_matrix.toarray())
log_vsm_corpus = pd.DataFrame(data=log_count_matrix, columns=vectorizer.get_feature_names_out())
new_kolom = df['Stopword Removal']
log_vsm_corpus.insert(0, 'Teks', new_kolom)
log_vsm_corpus
```

Kode tersebut merupakan bagian dari proses pengolahan teks yang melibatkan pembuatan representasi vektor dokumen menggunakan skema pembobotan Term Frequency (TF) dan Logarithmic Frequency. Dalam langkah pertama, `CountVectorizer` dari Scikit-learn digunakan untuk menghitung frekuensi kata-kata dalam teks, yang kemudian diubah menjadi matriks frekuensi kata. Selanjutnya, matriks tersebut diubah ke dalam skema Logarithmic Frequency dengan menghitung logaritma natural dari setiap elemen matriks. Hasilnya, DataFrame baru, `log_vsm_corpus`, dibuat dengan menambahkan kolom 'Teks' yang berisi teks asli sebelum penghapusan stopword. Representasi vektor dokumen ini dapat digunakan untuk analisis atau pemodelan teks dengan memberikan bobot yang lebih seimbang pada kata-kata yang muncul dalam jumlah yang bervariasi di dalam dokumen.

# **Binary Frequency**

Binary Frequency adalah salah satu metode pembobotan dalam pemodelan teks yang menyatakan kehadiran atau ketiadaan suatu kata dalam suatu dokumen tanpa memperhatikan berapa kali kata tersebut muncul. Dalam Binary Frequency, nilai pembobotan hanya mengindikasikan apakah suatu kata muncul atau tidak muncul dalam dokumen, dan tidak memperhitungkan berapa kali kata tersebut muncul. Jika suatu kata hadir dalam dokumen, bobotnya akan diatur menjadi 1, dan jika tidak hadir, bobotnya diatur menjadi 0.

Rumus Binary Frequency untuk suatu term $(t)$ dalam suatu dokumen $(d)$ dapat dinyatakan sebagai berikut :     


$ \text{Binary Frequency}(t, d) =
\begin{cases}
1 & \text{jika } t \text{ hadir dalam } d \\
0 & \text{jika } t \text{ tidak hadir dalam } d
\end{cases}
$

Metode ini sering digunakan dalam konteks di mana yang penting hanya mengetahui apakah suatu kata hadir atau tidak, tanpa memperhatikan seberapa sering kata tersebut muncul dalam dokumen. Binary Frequency sering digunakan sebagai langkah awal atau alternatif sederhana dalam representasi vektor dokumen.



```python
vectorizer = CountVectorizer(binary=True)
X = vectorizer.fit_transform(df['Stopword Removal'])
feature_names = vectorizer.get_feature_names_out()
corpus_vsm_binary = pd.DataFrame(data=X.toarray(), columns=feature_names)
kolom_baru = df['Stopword Removal']
corpus_vsm_binary.insert(0, 'Teks', kolom_baru)
corpus_vsm_binary
```

Kode tersebut merupakan bagian dari proses pembuatan representasi vektor dokumen dengan menggunakan skema Binary Frequency, di mana setiap elemen matriks menunjukkan kehadiran atau ketiadaan suatu kata dalam dokumen tanpa memperhitungkan frekuensinya. Dalam langkah pertama, CountVectorizer dari Scikit-learn diinisialisasi dengan parameter binary=True, mengindikasikan penggunaan Binary Frequency. Selanjutnya, matriks Binary Frequency diperoleh dengan memanggil fit_transform pada teks yang telah melalui tahap penghapusan stopword. Hasilnya, DataFrame baru, corpus_vsm_binary, dibuat dengan menambahkan kolom 'Teks' yang berisi teks asli sebelum penghapusan stopword.

# **TF-IDF**

TF-IDF (Term Frequency-Inverse Document Frequency) adalah metode pembobotan kata yang digunakan untuk mengevaluasi kepentingan suatu kata dalam suatu dokumen dalam konteks seluruh koleksi dokumen. Proses ini melibatkan dua konsep utama: Term Frequency (TF) dan Inverse Document Frequency (IDF).

**Term Frequency (TF)**

Term Frequency (TF) atau Frekuensi Kata mengukur seberapa sering sebuah kata muncul dalam sebuah dokumen. Konsep ini adalah bagian penting dari analisis teks dan pemrosesan bahasa alami. Term Frequency dapat membantu dalam menentukan seberapa penting sebuah kata dalam suatu dokumen.

Rumus Term Frequency (TF) pada umumnya dinyatakan sebagai jumlah kemunculan kata (term) dalam suatu dokumen dibagi dengan total jumlah kata dalam dokumen tersebut. Terdapat beberapa variasi rumus TF, dan salah satu yang umum digunakan adalah sebagai berikut:

$\text{TF}(t, d) = \frac{\text{jumlah kemunculan term } t \text{ dalam dokumen } d}{\text{total jumlah term dalam dokumen } d}$

**Inverse Document Frequency (IDF)**

Mengukur seberapa unik suatu kata dengan mengurangi dampak kata-kata yang muncul di banyak dokumen. IDF untuk suatu term $t$ dihitung sebagai logaritma dari rasio total dokumen dalam koleksi dibagi dengan jumlah dokumen yang mengandung term $t$ ditambah 1 untuk menghindari pembagian oleh nol.

$\text{IDF}(t) = \log\left(\frac{\text{total dokumen dalam koleksi}}{\text{jumlah dokumen yang mengandung term } t + 1}\right)$

**TF-IDF Score**

Bobot akhir suatu term dalam suatu dokumen diperoleh dengan mengalikan TF dan IDF.

$ \text{TF-IDF}(t, d) = \text{TF}(t, d) \times \text{IDF}(t) $

dimana :    

$\text{TF}(t, d) = \frac{\text{jumlah kemunculan term } t \text{ dalam dokumen } d}{\text{total kata dalam dokumen } d}$


$\text{IDF}(t) = \log\left(\frac{\text{total dokumen dalam koleksi}}{\text{jumlah dokumen yang mengandung term } t + 1}\right)$

Dengan metode ini, kata-kata yang muncul sering dalam suatu dokumen tetapi jarang muncul dalam koleksi secara keseluruhan diberi bobot yang lebih tinggi, sehingga menyoroti kepentingan relatif kata-kata tersebut dalam menggambarkan dokumen. TF-IDF digunakan dalam berbagai aplikasi, seperti sistem pengambilan informasi, klasifikasi teks, dan pencarian informasi.

# Latent Dirichlet Allocation (LDA)

Latent Dirichlet Allocation (LDA) adalah salah satu algoritma populer dalam topik modeling. Dikembangkan oleh David Blei, Andrew Ng, dan Michael Jordan pada tahun 2003, LDA memberikan kerangka kerja probabilistik untuk menggambarkan topik dalam suatu koleksi dokumen.

Konsep dasar di balik LDA adalah asumsi bahwa setiap dokumen adalah campuran atau kombinasi dari beberapa topik, dan setiap kata dalam dokumen berasal dari salah satu topik tersebut. Selanjutnya, algoritma LDA berusaha untuk mengidentifikasi distribusi topik yang mungkin menghasilkan dokumen-dokumen dalam koleksi tersebut.

Langkah-langkah umum dalam proses LDA:

1. **Inisialisasi:** Setiap kata dalam setiap dokumen diinisialisasi dengan suatu topik secara acak.
   
2. **Iterasi:** Selama beberapa iterasi, LDA mencoba memperbarui penugasan topik untuk setiap kata berdasarkan distribusi probabilitas.

3. **Perhitungan Distribusi Topik:** Algoritma LDA menghitung distribusi probabilitas topik untuk setiap dokumen dan distribusi probabilitas kata untuk setiap topik.

4. **Optimisasi:** Proses ini diulang hingga model konvergen, yaitu distribusi topik yang dihasilkan mencerminkan distribusi topik yang paling mungkin menghasilkan koleksi dokumen.

Hasil akhir dari proses LDA adalah representasi vektor topik untuk setiap dokumen dan distribusi kata dalam setiap topik. Dengan informasi ini, kita dapat mengidentifikasi topik utama dalam dokumen-dokumen tersebut.

LDA digunakan secara luas dalam berbagai bidang, termasuk analisis teks, temu kembali informasi, dan pengorganisasian konten. Meskipun LDA memberikan wawasan yang berharga, perlu diingat bahwa hasilnya sering kali memerlukan interpretasi manusia untuk memahami makna sebenarnya dari setiap topik yang dihasilkan.

Latent Dirichlet Allocation (LDA) adalah model probabilistik, dan rumusnya terdiri dari beberapa konsep kunci dalam statistika probabilitas dan teori distribusi. Berikut adalah rumus dasar yang digunakan dalam model LDA:



# 1. Proporsi topik dalam dokumen

Dalam konteks model Latent Dirichlet Allocation (LDA), proporsi topik dalam suatu dokumen diwakili oleh vektor \(\theta\), yang merupakan hasil dari distribusi Dirichlet. Proporsi topik dalam dokumen dapat dijelaskan dengan rumus sebagai berikut:

$ P(\theta | \alpha) = \text{Dirichlet}(\theta | \alpha)$

Di sini:
- $(\theta) $adalah vektor proporsi topik dalam suatu dokumen.
- $(alpha)$ adalah parameter distribusi Dirichlet yang mengontrol seberapa banyak setiap topik dapat berkontribusi pada dokumen. Nilai $(alpha)$ yang tinggi cenderung menghasilkan dokumen dengan proporsi topik yang merata, sedangkan nilai $(alpha)$ yang rendah dapat menghasilkan dokumen dengan fokus pada beberapa topik tertentu.

Dengan distribusi Dirichlet ini, setiap nilai $\theta$ dalam vektor $\theta$ menunjukkan seberapa besar kontribusi topik ke-$(k)$ terhadap dokumen tersebut. Secara matematis, nilai-nilai $(\theta_k)$ untuk $(k)$ dari $(1)$ hingga $(K)$ (jumlah total topik) akan membentuk suatu distribusi probabilitas yang jumlah totalnya sama dengan $(1)$.

```python
data = pd.DataFrame(corpus_vsm)

X = data.drop('Teks', axis=1)

from sklearn.decomposition import LatentDirichletAllocation

topik = 3
alpha = 0.1 #distribusi topik dalam dokumen
beta = 0.2 #distribusi kata dalam topik

#membuat model LDA
lda = LatentDirichletAllocation(n_components=topik, doc_topic_prior=alpha, topic_word_prior=beta)
proporsi_topik_dokumen = lda.fit_transform(X)

dokumen = data['Teks']
output_proporsi_TD = pd.DataFrame(proporsi_topik_dokumen, columns=['Topik 1', 'Topik 2', 'Topik 3'])
output_proporsi_TD.insert(0,'Teks', dokumen)
output_proporsi_TD
```

Dalam blok kode tersebut, data yang awalnya berupa matriks term frequency hasil dari representasi vektor ruang tematik (VSM) diubah menjadi bentuk DataFrame menggunakan pandas. Kolom 'Teks' dihapus dari DataFrame, dan representasi dokumen dalam bentuk matriks numerik disiapkan untuk dianalisis menggunakan model Latent Dirichlet Allocation (LDA). Sebuah model LDA dengan tiga topik ditentukan dengan parameter distribusi topik dalam dokumen $\alpha$ dan distribusi kata dalam topik $\beta$. Setelah model LDA dilatih pada data, proporsi topik dalam setiap dokumen diekstrak. Hasilnya direpresentasikan dalam DataFrame baru, yang mencakup kolom 'Teks' dan proporsi topik untuk setiap dokumen.

# 2. Distribusi kata pada topik

Distribusi kata dalam suatu topik dalam konteks model Latent Dirichlet Allocation (LDA) menggambarkan seberapa besar probabilitas setiap kata muncul dalam topik tersebut. Distribusi ini direpresentasikan oleh vektor $(\beta_k)$, di mana $(k)$ adalah indeks topik tertentu. Setiap elemen $(\beta_{k,v})$ dari vektor tersebut menyatakan probabilitas kata ke-$(v)$ muncul dalam topik $(k)$.

Rumus distribusi kata dalam suatu topik (\(\beta\)) dapat dirumuskan sebagai berikut:

$P(w_{d,n} = v | z_{d,n} = k, \beta_k) = \beta_{k,v} $

Di sini:
- $(w_{d,n})$ adalah kata ke-$(n)$ dalam dokumen $(d)$.
- $(z_{d,n} = k)$ menyatakan bahwa kata tersebut ditugaskan ke topik $(k)$.
- $(\beta_{k,v})$ adalah probabilitas kata ke-$(v)$ dalam topik $(k)$.

Rumus ini memberikan gambaran tentang seberapa kuat keterkaitan antara suatu kata dengan suatu topik tertentu. Nilai-nilai $(\beta_{k,v})$ yang tinggi menunjukkan bahwa kata tersebut sangat mungkin muncul dalam topik $(k)$, sementara nilai-nilai yang rendah menunjukkan probabilitas yang lebih rendah. Distribusi kata dalam topik menjadi kunci dalam memahami makna dan konten setiap topik dalam model LDA.

```python
distribusi_kata_topik = pd.DataFrame(lda.components_)
distribusi_kata_topik
```

Blok kode "distribusi_kata_topik = pd.DataFrame(lda.components_)" menghasilkan DataFrame yang menyajikan distribusi kata dalam setiap topik dari model Latent Dirichlet Allocation (LDA). Setiap baris dalam DataFrame mencerminkan suatu topik, sedangkan setiap kolom mewakili kata-kata dalam korpus. Nilai di setiap sel mencerminkan bobot relatif kata dalam topik tersebut. Analisis distribusi kata dalam topik memberikan pemahaman tentang kata-kata yang paling mencirikan suatu topik, membantu interpretasi dan makna dari setiap topik dalam model LDA.

Teks summarization adalah suatu metode komputasional yang bertujuan untuk membuat ringkasan atau cuplikan singkat dari teks yang lebih panjang atau kompleks. #**TF-IDF**

TF-IDF (Term Frequency-Inverse Document Frequency) adalah metode statistik yang digunakan dalam teks summarization untuk menilai kepentingan relatif suatu kata dalam suatu dokumen atau korpus. Rumus TF-IDF menggabungkan Term Frequency (TF), yang mengukur seberapa sering suatu kata muncul dalam dokumen, dengan Inverse Document Frequency (IDF), yang mempertimbangkan seberapa umum atau unik kata tersebut dalam seluruh korpus. Penggunaan TF-IDF dalam teks summarization memberikan bobot pada kata-kata berdasarkan keunikannya dan membantu mengidentifikasi kata-kata kunci serta menekankan informasi penting. Selain itu, TF-IDF dapat digunakan untuk mereduksi dimensi dalam representasi teks, memilih fitur-fitur informatif, dan menyediakan landasan bagi ekstraksi informasi kunci dalam pembuatan ringkasan. Dengan memperhatikan frekuensi dan distribusi kata-kata, TF-IDF membantu meningkatkan kualitas representasi teks, memfasilitasi proses ekstraksi informasi, dan menyokong tugas teks summarization secara efektif.

berikut code untuk tf idf

```python
import math
from sklearn.feature_extraction.text import TfidfVectorizer

df_tampilan = pd.DataFrame(df_tampilan)
# Hitung TF untuk setiap kata dalam setiap kalimat
tf_values = []
for index, row in df_tampilan.iterrows():
    tf_dict = {}
    for word in row['kalimat'].split():
        tf_dict[word] = row['kalimat'].split().count(word) / row['jumlah_kata']
    tf_values.append(tf_dict)

# Menghitung IDF untuk setiap kata
idf_values = {}
total_documents = len(df_tampilan)
for tf in tf_values:
    for word in tf:
        if word in idf_values:
            idf_values[word] += 1
        else:
            idf_values[word] = 1

# Menghitung TF-IDF
tfidf_values = []
for tf in tf_values:
    tfidf_dict = {}
    for word, tf_value in tf.items():
        tfidf_dict[word] = tf_value * math.log(total_documents / (1 + idf_values[word]))
    tfidf_values.append(tfidf_dict)

# Konversi list of dicts ke dalam DataFrame
df_tfidf = pd.DataFrame(tfidf_values)
df_tfidf.fillna(0, inplace=True)

# Tampilkan hasil TF-IDF DataFrame
df_tfidf
```

# **Cosine Similarity**

Cosine similarity adalah metrik yang digunakan dalam teks summarization untuk mengukur sejauh mana dua vektor teks mirip berdasarkan sudut cosine antara mereka. Dalam konteks ini, vektor teks direpresentasikan oleh nilai TF-IDF (Term Frequency-Inverse Document Frequency) dari kata-kata dalam dokumen atau kalimat. Cosine similarity mengukur kesamaan antara dua vektor dengan menghitung cosinus dari sudut yang terbentuk oleh vektor-vektor tersebut dalam ruang vektor. Nilai cosine similarity berkisar antara 0 (vektor sepenuhnya ortogonal, tidak ada kesamaan) hingga 1 (vektor sepenuhnya sejajar, kesamaan maksimum). Dalam konteks summarization, cosine similarity digunakan untuk mengevaluasi seberapa mirip suatu kalimat atau dokumen dengan kalimat atau dokumen lainnya. Metrik ini memungkinkan model untuk menentukan tingkat relevansi atau kemiripan antara teks, dan hasilnya dapat digunakan untuk memilih atau mengurutkan kalimat-kalimat yang paling relevan dalam proses pembuatan ringkasan. Semakin tinggi nilai cosine similarity, semakin mirip atau relevan kedua vektor tersebut, yang dapat menghasilkan ringkasan yang lebih akurat dan informatif.

Agar lebih mudah memahami perhitungan cosine similarity perhatikan hal berikut :

Bayangkan kita memiliki dua daftar kata-kata yang mewakili dua kalimat atau dokumen. Untuk mengukur seberapa mirip keduanya, kita menggunakan cosine similarity. Langkah-langkahnya adalah sebagai berikut:

1. **Hitung Jumlah Kalikan:**
   - Hitung hasil penjumlahan dari perkalian setiap kata pada posisi yang sama dari kedua daftar kata.

2. **Hitung Panjang (Magnitude):**
   - Hitung panjang (magnitude) dari masing-masing daftar kata.
   - Panjangnya dihitung dengan menjumlahkan kuadrat dari setiap kata dan mengakarkan hasilnya.

3. **Hitung Cosine Similarity:**
   - Bagi hasil penjumlahan kalikan tadi dengan hasil perkalian panjang kedua daftar kata.
   - Ini memberikan nilai cosine similarity antara 0 (tidak mirip) dan 1 (sangat mirip).

Contoh:
- Daftar kata pertama: [3, 1, 2]
- Daftar kata kedua: [1, 2, 0]

1. Hitung Jumlah Kalikan: $ (3 \times 1) + (1 \times 2) + (2 \times 0) = 5 $
2. Hitung Panjang:
   - Panjang Daftar Pertama: $ \sqrt{3^2 + 1^2 + 2^2} = \sqrt{14} $
   - Panjang Daftar Kedua: $ \sqrt{1^2 + 2^2 + 0^2} = \sqrt{5} $
3. Hitung Cosine Similarity: $ \frac{5}{\sqrt{14} \times \sqrt{5}} \approx 0.75 $

Hasilnya adalah sekitar 0.75, yang berarti kedua daftar kata tersebut cukup mirip. Semakin mendekati 1, semakin mirip keduanya.

# **Page Rank**

PageRank adalah algoritma yang digunakan oleh mesin pencari, terutama oleh Google, untuk menilai dan mengurutkan halaman web berdasarkan tingkat otoritas dan relevansinya. Konsepnya mirip dengan cara kita memberi peringkat populeritas atau pentingnya halaman web. Bayangkan setiap halaman web sebagai orang dalam sebuah jaringan sosial. Jika banyak orang (halaman web lain) menghubungi atau memberi suara pada seseorang (halaman web), orang itu dianggap penting. Dengan PageRank, setiap halaman web memberikan "suara" dalam bentuk tautan ke halaman lain. Semakin banyak tautan masuk yang dimiliki halaman web, semakin tinggi peringkatnya. Tetapi tidak semua suara dianggap sama. Suara dari halaman yang memiliki peringkat tinggi lebih berharga daripada suara dari halaman dengan peringkat rendah. Dengan cara ini, PageRank membantu mesin pencari menentukan halaman web mana yang paling penting dan relevan, sehingga hasil pencarian yang diberikan kepada pengguna menjadi lebih baik.

untuk lebih memahami page rank, perhatikan perhitungan berikut :    

PageRank dihitung menggunakan algoritma yang mempertimbangkan jumlah dan kualitas tautan yang mengarah ke suatu halaman web. Secara matematis, rumus PageRank dapat dijelaskan sebagai berikut:

$ \text{PR}(A) = (1 - d) + d \left( \frac{\text{PR}(B)}{L(B)} + \frac{\text{PR}(C)}{L(C)} + \ldots + \frac{\text{PR}(N)}{L(N)} \right) $

Di sini:
- $\text{PR}(A)$ adalah PageRank dari halaman web A.
- $(d)$ adalah faktor damping (biasanya diatur sekitar 0,85), yang merepresentasikan probabilitas bahwa pengguna akan mengikuti tautan daripada menavigasi ke halaman acak.
- $\text{PR}(B)$, $\text{PR}(C)$, ..., $\text{PR}(N)$ adalah PageRank dari halaman-halaman yang terhubung ke halaman A.
- $L(B)$, $L(C)$, ..., $L(N)$ adalah jumlah tautan keluar dari halaman-halaman B, C, ..., N.

Mari kita ilustrasikan ini dengan contoh sederhana:

1. Ada tiga halaman web: A, B, dan C.
2. Semua halaman memiliki PageRank awal yang sama (misalnya, 1/3).
3. Halaman B dan C memiliki tautan ke halaman A.

Mari kita hitung PageRank untuk halaman A:

$ \text{PR}(A) = (1 - 0.85) + 0.85 \left( \frac{1/3}{1} + \frac{1/3}{1} \right) $

$\text{PR}(A) = 0.15 + 0.85 \left( \frac{2/3}{1} \right) $

$ \text{PR}(A) = 0.15 + 0.85 \times \frac{2}{3} $

$ \text{PR}(A) = 0.15 + 0.566666... $

$ \text{PR}(A) \approx 0.716 $

Ini adalah perhitungan sederhana untuk mengilustrasikan konsep PageRank. Dalam praktiknya, algoritma PageRank diiterasi secara berulang hingga konvergensi untuk mendapatkan nilai akhir PageRank yang lebih akurat.

# **Closseness Centrality**



Closeness centrality adalah salah satu metrik sentralitas dalam jaringan yang digunakan untuk mengukur seberapa dekat suatu node dengan node lainnya dalam graf. Dalam konteks teks summarization, closeness centrality dapat diaplikasikan pada representasi graf dari hubungan antara kalimat-kalimat dalam suatu teks. Metrik ini memberikan informasi tentang seberapa mudah suatu kalimat dapat diakses oleh kalimat-kalimat lain dalam teks tersebut.

Dalam implementasi kode yang Anda berikan, closeness centrality dihitung untuk setiap kalimat dalam graf representasi hubungan kalimat (misalnya, graf kalimat yang terhubung berdasarkan kosinus similarity). Kalimat yang memiliki closeness centrality tinggi cenderung menjadi pusat komunikasi dalam teks, karena lebih dekat dengan kalimat-kalimat lainnya.

Dalam konteks teks summarization, closeness centrality dapat membantu mengidentifikasi kalimat-kalimat yang secara esensial "menghubungkan" atau merepresentasikan inti informasi dalam teks. Kalimat-kalimat ini memiliki potensi untuk menjadi penting dalam proses pembuatan ringkasan karena mereka memiliki keterkaitan yang erat dengan kalimat-kalimat lainnya. Memilih kalimat-kalimat dengan closeness centrality tinggi dapat membantu dalam mengekstraksi inti informasi atau gagasan utama dari teks, yang merupakan langkah kritis dalam proses teks summarization. Oleh karena itu, closeness centrality menjadi salah satu faktor yang dapat digunakan untuk menentukan relevansi dan kontribusi kalimat dalam memahami struktur dan makna keseluruhan teks.

Closeness centrality mengukur seberapa dekat suatu node (kalimat dalam konteks teks summarization) dengan semua node lainnya dalam graf. Rumusnya dapat dijelaskan sebagai:

$ C(x) = \frac{1}{\sum_{y \neq x} d(x, y)} $

Di sini:
- $ C(x) $ adalah closeness centrality dari node $ x $ (kalimat).
- $ d(x, y) $ adalah jarak terpendek (misalnya, jumlah tautan atau kesamaan kosinus) antara node $ x $ dan node $ y $.

Closeness centrality diperoleh dengan membalikkan jumlah jarak terpendek dari node $ x $ ke semua node lainnya. Semakin kecil nilai $ \sum_{y \neq x} d(x, y) $, semakin besar nilai closeness centrality-nya. Dalam konteks teks summarization, kalimat-kalimat dengan closeness centrality yang tinggi cenderung menjadi pusat atau penting dalam merangkum informasi karena dekat dengan kalimat-kalimat lainnya dalam teks.

## **Eigenvector Centrality**

Eigenvector Centrality adalah metrik sentralitas yang digunakan dalam analisis jaringan untuk mengukur seberapa penting atau sentral suatu node dalam konteks keterhubungan dengan node-node yang juga memiliki sentralitas tinggi. Dalam teks summarization, Eigenvector Centrality dapat diaplikasikan pada representasi graf hubungan antar kalimat atau dokumen.

Rumus Eigenvector Centrality untuk suatu node $v$ dalam graf adalah sebagai berikut:

$ C(v) = \frac{1}{\lambda} \sum_{u \in N(v)} A(u, v) \cdot C(u) $

Di sini:
- $ C(v) $ adalah Eigenvector Centrality dari node $v$.
- $ N(v) $ adalah himpunan tetangga dari node $v$, yaitu node-node yang terhubung langsung dengan $v$.
- $ A(u, v) $ adalah elemen matriks ketetanggaan yang menunjukkan apakah terdapat sisi (edge) antara node $u$ dan $v$.
- $ {λ} $ adalah nilai eigenvalue utama (dominan) dari matriks ketetanggaan.

Dalam konteks teks summarization, Eigenvector Centrality dapat membantu mengidentifikasi kalimat-kalimat atau dokumen-dokumen yang secara struktural penting dalam jaringan kalimat atau dokumen. Kalimat atau dokumen dengan Eigenvector Centrality yang tinggi cenderung menjadi pusat atau memiliki keterkaitan yang kuat dengan kalimat atau dokumen lain yang juga dianggap penting. Dengan memilih kalimat-kalimat atau dokumen-dokumen ini, kita dapat fokus pada inti informasi atau gagasan utama dalam teks saat melakukan proses summarization. Sehingga, Eigenvector Centrality menjadi alat yang bermanfaat untuk memahami dan mengekstrak elemen-elemen krusial dalam teks.

   \begin{matrix}
   & \text{1} & \text{2} & \text{3} \\
   \text{1} & - & 1 & 1 \\
   \text{2} & 1 & - & 1 \\
   \text{3} & 1 & 1 & -
   \end{matrix}
   $

   Di sini, tanda $1$ menunjukkan adanya hubungan atau kesamaan antara dua kalimat.

2. **Hitung Eigenvector Centrality:**
   - Dengan menggunakan rumus Eigenvector Centrality, kita lakukan iterasi untuk menghitung sentralitas eigenvector untuk setiap kalimat.

   $ C(1) = \frac{1}{\lambda} \times (1 \times C(2) + 1 \times C(3)) $
   $ C(2) = \frac{1}{\lambda} \times (1 \times C(1) + 1 \times C(3)) $
   $ C(3) = \frac{1}{\lambda} \times (1 \times C(1) + 1 \times C(2)) $

   Iterasi ini dilakukan hingga nilai Eigenvector Centrality konvergen.

3. **Normalisasi Nilai:**

# **C. EKSTRAKSI KATA KUNCI**

Ekstraksi kata kunci adalah proses identifikasi dan pengekstrakan kata-kata atau frasa yang paling mewakili atau mencerminkan inti informasi dari suatu teks atau dokumen. Tujuan dari ekstraksi kata kunci adalah untuk menyajikan rangkuman singkat dari isu atau topik utama yang dibahas dalam teks, membantu pembaca atau sistem informasi untuk dengan cepat memahami esensi dari dokumen tersebut.

Terdapat beberapa metode yang dapat digunakan untuk melakukan ekstraksi kata kunci, antara lain:

1. **Frekuensi Kata:**
   - Menghitung frekuensi kemunculan setiap kata dalam teks dan mengekstrak kata-kata dengan frekuensi tertinggi. Kata-kata yang muncul lebih sering dianggap lebih penting.

2. **TF-IDF (Term Frequency-Inverse Document Frequency):**
   - Mengukur seberapa penting suatu kata dalam suatu dokumen relatif terhadap frekuensinya di seluruh koleksi dokumen. Kata-kata dengan nilai TF-IDF tinggi cenderung menjadi kata kunci.

3. **TextRank:**
   - Mengadaptasi konsep algoritma PageRank dari dunia jaringan untuk menilai pentingnya kata-kata dalam suatu teks berdasarkan hubungan dan kedekatan antar kata-kata tersebut.

4. **N-gram:**
   - Mengekstrak serangkaian kata berurutan (n-gram) yang paling umum atau penting dalam teks.

5. **Deep Learning:**
   - Menggunakan model-model deep learning, seperti BERT atau GPT, untuk menghasilkan representasi teks yang kompleks dan mengekstrak kata kunci berdasarkan pemahaman yang lebih kontekstual.

Proses ekstraksi kata kunci penting dalam berbagai aplikasi, termasuk dalam pemrosesan bahasa alami (NLP), analisis teks, dan pembuatan ringkasan otomatis. Hasil ekstraksi kata kunci dapat digunakan untuk memahami isu utama dalam suatu teks, meningkatkan pencarian informasi, dan menyusun ringkasan teks dengan cepat.

selain itu untuk melakukan ekstraksi kata kunci bisa menggunakan Rake.
RAKE, singkatan dari Rapid Automatic Keyword Extraction, adalah salah satu metode ekstraksi kata kunci yang dirancang untuk mengekstrak kata kunci secara cepat dan efisien dari sebuah teks. Metode ini menggunakan pendekatan berbasis frekuensi dan kombinasi kata-kata yang membentuk frasa kandidat.

Berikut adalah langkah-langkah umum dalam algoritma RAKE:

1. **Praproses Teks:**
   - Tahap awal melibatkan praproses teks, termasuk penghilangan tanda baca, konversi teks menjadi huruf kecil, dan langkah-langkah pembersihan lainnya.

2. **Pembentukan Frasa Kandidat:**
   - Teks dibagi menjadi frasa-frasa kandidat berdasarkan pemisah kata. Frasa-frasa ini dapat terdiri dari satu kata atau lebih.

3. **Hitung Skor Frekuensi dan Frekuensi Kata-Kata Kunci:**
   - Setiap frasa kandidat diberi skor berdasarkan frekuensinya dalam teks dan jumlah kata-kata kunci yang terkandung di dalamnya. Skor ini membantu dalam menilai potensi kata-kata kunci.

4. **Hitung Skor RAKE:**
   - Skor RAKE dihitung dengan membagi skor frekuensi frasa dengan jumlah kata dalam frasa tersebut. Ini membantu mengidentifikasi frasa yang memiliki kata-kata kunci dan memiliki panjang yang sesuai.

5. **Urutkan dan Pilih Kata Kunci:**
   - Frasa-frasa kandidat diurutkan berdasarkan skor RAKE, dan kata kunci yang dihasilkan dipilih dari urutan teratas.

6. **Hasilkan Kata Kunci Final:**
   - Kata kunci final diperoleh dari pemilihan frasa-frasa kunci yang memenuhi kriteria tertentu, seperti memiliki skor RAKE di atas ambang batas tertentu.

RAKE sering digunakan dalam konteks di mana ekstraksi kata kunci yang cepat diperlukan, seperti dalam pengindeksan dokumen atau pemrosesan teks besar. Meskipun metode ini relatif sederhana, ia dapat memberikan hasil yang baik dalam banyak situasi, terutama ketika teks memiliki struktur frasa yang kuat.

Pertama-tama, mari kita ambil contoh kalimat-kalimat berikut untuk menjelaskan secara manual bagaimana RAKE dapat digunakan untuk menghitung kata kunci:

1. "Hewan peliharaan adalah teman yang baik."
2. "Anjing dan kucing adalah contoh hewan peliharaan."
3. "Setiap orang harus memberi makan dan merawat hewan peliharaan dengan baik."

Langkah-langkah manual menggunakan RAKE:

1. **Praproses Teks:**
   - Tidak perlu melakukan praproses pada contoh ini karena teksnya sederhana.

2. **Pembentukan Frasa Kandidat:**
   - Frasa kandidat dibentuk dengan membagi teks berdasarkan pemisah kata. Misalnya:
     - "Hewan", "peliharaan", "teman", "baik", dll.

3. **Hitung Skor Frekuensi dan Frekuensi Kata-Kata Kunci:**
   - Hitung frekuensi setiap frasa dan jumlah kata kunci yang terkandung di dalamnya. Misalnya:
     - "Hewan": frekuensi = 3, jumlah kata kunci = 1
     - "peliharaan": frekuensi = 3, jumlah kata kunci = 2
     - ...

4. **Hitung Skor RAKE:**
   - Hitung skor RAKE dengan membagi skor frekuensi frasa dengan jumlah kata dalam frasa tersebut. Misalnya:
     - RAKE("Hewan") = frekuensi / jumlah_kata = 3 / 1 = 3
     - RAKE("peliharaan") = frekuensi / jumlah_kata = 3 / 2 = 1.5
     - ...

5. **Urutkan dan Pilih Kata Kunci:**
   - Urutkan frasa-frasa berdasarkan skor RAKE secara menurun.
     - Urutan teratas mungkin: "Hewan", "peliharaan", "teman", dll.

6. **Hasilkan Kata Kunci Final:**
   - Pilih kata kunci dari urutan teratas berdasarkan ambang batas tertentu. Misalnya, ambang batas skor RAKE > 2.
     - Kata kunci final: "Hewan", "peliharaan"

Dengan langkah-langkah ini, kita dapat secara manual menggunakan algoritma RAKE untuk mengidentifikasi kata kunci dari teks. Proses ini dapat diulangi untuk teks yang lebih besar dengan langkah-langkah yang serupa, dan kata kunci dapat dipilih sesuai dengan ambang batas atau kriteria tertentu.
